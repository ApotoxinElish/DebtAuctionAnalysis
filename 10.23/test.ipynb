```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("prices_jp/prices.dai.2021.by_minute.20220221.csv")
data.head()
```

```python
data.date_time = pd.to_datetime(data.date_time, format="%Y/%m/%d %H:%M:%S")
data.info()
```

```python
time = pd.Timestamp("2021/5/6  19:09:26")
time
delta = pd.Timedelta("3 hours")
```

```python
usefuldata = data.loc[(time - delta < data.date_time) & (data.date_time < time + delta)]

usefuldata
```

```python
print(usefuldata.dai.mean())

usefuldata.dai.var()
```

```python
plt.figure()
plt.scatter(
    usefuldata.date_time,
    usefuldata.dai,
    s=20,
    edgecolor="black",
    c="darkorange",
    label="dai",
)
plt.xlabel("date_time")
plt.ylabel("dai")
plt.title("2021 dai")
plt.legend()
plt.show()
```

```python
dai = list(usefuldata.dai)
dai = [round(i, 3) for i in dai]
len(dai)
dai
```

```python
dai_x = list(set(dai))
dai_x.sort()
len(dai_x)
dai_x
```

```python
count_y = [dai.count(i) for i in dai_x]
count_y
```

```python
plt.plot(dai_x, count_y, color="cornflowerblue", label="max_depth=2", linewidth=2)
```
